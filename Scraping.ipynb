```python
# Import library
import pandas as pd
from google_play_scraper import reviews, Sort

# Fungsi scraping review dari Google Play
def scrape_reviews(app_id, total=10000):
    all_reviews = []
    count = 0

    while len(all_reviews) < total:
        batch, _ = reviews(
            app_id,
            lang='en',
            country='us',
            sort=Sort.NEWEST,
            count=min(200, total - len(all_reviews)),
            filter_score_with=None
        )
        all_reviews.extend(batch)
        count += 1
        print(f"Batch {count} - Collected {len(all_reviews)} reviews")

    df = pd.DataFrame(all_reviews)
    return df[['content', 'score']]

# Contoh scraping aplikasi TikTok
app_reviews = scrape_reviews('com.zhiliaoapp.musically', total=100000)

# Simpan ke CSV
app_reviews.to_csv('tiktok_reviews.csv', index=False)
print("Scraping selesai. Data disimpan di 'tiktok_reviews.csv'")
```
